```c++
   Long64_t nentries = fch->GetEntries();
   for (Long64_t jentry=0; jentry<nentries;jentry+=1) {
      fch->GetEntry(jentry);
      if(jentry%1000000==0)cout<<jentry<<'\n';
      for (int ch = 0; ch < 32; ch++) {
        //h[ch]->Fill(chdata2[ch]);
        if(ch<16&&chdata2[ch]>pedel[ch]){
            hc[ch]->Fill((chdata2[ch]*k_l[ch][0]+k_l[ch][1])/1000.);
        }
      if(ch<16&&chdata3[ch]>pedelt[ch]){
            htc[ch]->Fill((chdata3[ch]*k_lt[ch][0]+k_lt[ch][1])/1000.);
        }
      }
      
       
/********hit 结构 35°**************/
    // 重置hit计数器
    nhits = 0;
    for (int i = 0; i < 5; i++) {
        xe[i] = 0;
        xce[i] = 0;
        xs[i] = -1;
        xe2=0;
    }       
    // 临时存储hit数据
    Double_t tmp_xe[16] = {0};      // 原始能量
    Double_t tmp_xce[16] = {0};     // 刻度后能量
    Int_t tmp_xs[16] = {-1};        // 通道号
    Int_t tmp_nhits = 0;            // 临时hit计数器
       
     // 收集所有超过阈值的hit
    for (int ch = 0; ch < 16; ch++) {  // 只处理0-15通道
        if (chdata2[ch] > pedel[ch]) {
            tmp_xe[tmp_nhits] = chdata2[ch];
            tmp_xce[tmp_nhits] = (chdata2[ch] * k_l[ch][0] + k_l[ch][1]) / 1000.0;
            tmp_xs[tmp_nhits] = ch;
            tmp_nhits++;
        }
    }
    
    // 记录大硅信号（通道17）
    xe2 = chdata2[17];
    
    // 按能量从大到小排序（选择排序）
    nhits = tmp_nhits;
    for (int i = 0; i < min(5, nhits); i++) {
        // 找出当前最大能量的hit
        int max_index = i;
        for (int j = i + 1; j < nhits; j++) {
            if (tmp_xe[j] > tmp_xe[max_index]) {
                max_index = j;
            }
        }
        
        // 保存前5个hit
        xe[i] = tmp_xe[max_index];
        xce[i] = tmp_xce[max_index];
        xs[i] = tmp_xs[max_index];      
        // 将已选中的hit移到末尾（避免重复选择）
        if (max_index != i) {
            swap(tmp_xe[max_index], tmp_xe[i]);
            swap(tmp_xce[max_index], tmp_xce[i]);
            swap(tmp_xs[max_index], tmp_xs[i]);
        }
    }

/********hit 结构 90°**************/

    // 重置hit计数器
    tnhits = 0;
    for (int i = 0; i < 5; i++) {
        txe[i] = 0;
        txce[i] = 0;
        txs[i] = -1;
        txe2=0;
        txe3=0;
    }       
    // 临时存储hit数据
    Double_t ttmp_xe[16] = {0};      // 原始能量
    Double_t ttmp_xce[16] = {0};     // 刻度后能量
    Int_t ttmp_xs[16] = {-1};        // 通道号
    Int_t ttmp_nhits = 0;            // 临时hit计数器
       
     // 收集所有超过阈值的hit
    for (int ch = 0; ch < 16; ch++) {  // 只处理0-15通道
        if (chdata3[ch] > pedelt[ch]) {
            ttmp_xe[ttmp_nhits] = chdata3[ch];
            ttmp_xce[ttmp_nhits] = (chdata3[ch] * k_lt[ch][0] + k_lt[ch][1]) / 1000.0;
            ttmp_xs[ttmp_nhits] = ch;
            ttmp_nhits++;
        }
    }
    
    // 记录大硅信号（通道16,17）
    txe2 = chdata3[16];
    txe3 = chdata3[17];
    
    // 按能量从大到小排序（选择排序）
    tnhits = ttmp_nhits;
    for (int i = 0; i < min(5, tnhits); i++) {
        // 找出当前最大能量的hit
        int tmax_index = i;
        for (int j = i + 1; j < tnhits; j++) {
            if (ttmp_xe[j] > ttmp_xe[tmax_index]) {
                tmax_index = j;
            }
        }
        
        // 保存前5个hit
        txe[i] = ttmp_xe[tmax_index];
        txce[i] = ttmp_xce[tmax_index];
        txs[i] = ttmp_xs[tmax_index];      
        // 将已选中的hit移到末尾（避免重复选择）
        if (tmax_index != i) {
            swap(ttmp_xe[tmax_index], ttmp_xe[i]);
            swap(ttmp_xce[tmax_index], ttmp_xce[i]);
            swap(ttmp_xs[tmax_index], ttmp_xs[i]);
        }
    }


    // 填充树
    hitTree->Fill();
       
   }
```
