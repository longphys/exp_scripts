```c++
Double_t fitGaussian(TH1F *h[809][4],Int_t runnum,Int_t ii) {
    // 定义高斯拟合函数
    TF1* gaussian = new TF1("gaussian", "gaus", _minfit, _maxfit);
    h[runnum][ii]->Fit(gaussian, "RQ");

    // 获取拟合结果中的均值
    return gaussian->GetParameter(1);
}
```
